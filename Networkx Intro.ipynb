````python
import networkx as nx
````

````python
G = nx.Graph()
G
````

# Nodes

````python
G.add_node(1)
G.add_node(6, color='brown')
G.add_nodes_from([2, 3])
G.add_nodes_from([
    (4, {'color': 'red'}),
    (5, {'color': 'green'})
])


nx.draw(G, with_labels=True)
````

````python
H = nx.path_graph(10)
nx.draw(H, with_labels=True)
H
````

# Edges

````python
G = nx.Graph()
G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)
G.add_edges_from([
    (4, 2, {'weight': 3.14}),
    (3, 5)
])
````

````python
nx.draw(G, with_labels=True)
````
