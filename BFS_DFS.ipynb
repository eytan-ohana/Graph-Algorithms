# DFS

```python
G = nx.gnm_random_graph(12, 7)
G = nx.complete_graph(8)
G.remove_edges_from([(0, 1), (1, 2), (6, 3), (6, 5), (7, 0), (6, 0),
                     (2, 5), (4, 1), (7, 3), (3, 4), (7, 2), (5, 3)])

for node in G:
    G.nodes[node]['marked'] = 'False'
    print(node, G.nodes[node])
    
nx.draw(G, with_labels=True)
```

```python
for node, a in G.nodes.items():
    print(node, a)
```

```python
for e, a in G.edges.items():
    a['marked'] = False
```
