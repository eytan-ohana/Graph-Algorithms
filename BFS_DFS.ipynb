```python
G = nx.gnm_random_graph(12, 7)
G = nx.complete_graph(8)
G.remove_edges_from([(0, 1), (1, 2), (6, 3), (6, 5), (7, 0), (6, 0),
                     (2, 5), (4, 1), (7, 3), (3, 4), (7, 2), (5, 3)])

for node in G.nodes:
    G.nodes[node]['marked'] = False
for edge in G.edges:
    G.edges[edge]['marked'] = False
    
    
draw_graph(G)
```
